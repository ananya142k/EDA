```python
ar_2000_to_2010 = ar[(ar['Year'] >= 2000) & (ar['Year'] <= 2010)]

Energy_stats = ar_2000_to_2010 ['Energy'].describe()
print(Energy_stats)
print('')
Speechiness_stats = ar_2000_to_2010 ['Speechiness'].describe()
print(Speechiness_stats)
print('')
Acousticness_stats = ar_2000_to_2010 ['Acousticness'].describe()
print(Acousticness_stats)
print('')
Valence_stats = ar_2000_to_2010 ['Valence'].describe()
print(Valence_stats)
print('')
Tempo_stats = ar_2000_to_2010 ['Tempo'].describe()
print(Tempo_stats)
print('')
```

```python
import matplotlib.pyplot as plt

ar_2000_to_2010 = ar[(ar['Year'] >= 2000) & (ar['Year'] <= 2010)]

ar_audio_features = ['Energy', 'Speechiness', 'Acousticness', 'Valence', 'Tempo']

for feature in ar_audio_features:
    plt.figure(figsize=(8, 6))
    ar_2000_to_2010[feature].plot(kind='kde')
    plt.title(f'{feature} (2000-2010)')
    plt.xlabel('Distributions')
    plt.ylabel('Frequency')
    plt.show()
```

## ***Skewness***

```python
import pandas as pd
from scipy.stats import skew

ar_2000_to_2010 = ar[(ar['Year'] >= 2000) & (ar['Year'] <= 2010)]

audio_features = ['Energy', 'Speechiness', 'Acousticness', 'Valence', 'Tempo']

skewness_data = {}

for feature in audio_features:
    skewness = skew(ar_2000_to_2010[feature].dropna())
    skewness_data[feature] = skewness

skewness_df = pd.DataFrame(list(skewness_data.items()), columns=['Feature', 'Skewness'])

print(skewness_df)
```
